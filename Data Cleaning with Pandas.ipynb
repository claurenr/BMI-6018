```python
# -*- coding: utf-8 -*-

#%% Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#%% Importing Data
flights_data = pd.read_csv('flights.csv')
flights_data.head(10)
weather_data_pd = pd.read_csv('weather.csv')
weather_data_np = weather_data_pd.to_numpy()
```

```python
#Question 1 How many flights were there from JFK to SLC? Int q_1 
#Filter the data to flights that flew from JFK to SLC
new_df = flights_data.loc[(flights_data['origin'] == "JFK") & (flights_data['dest'] == "SLC")]

#Shape returns the number of rows which equals the number of flights
q_1 = new_df.shape[0]
q_1
```

```python
#Question 2 How many airlines fly to SLC? Should be intq_2 
#group by flight destination
dest_grp = flights_data.groupby(['dest'])
#Then select just SLC from the group by destination data
dest_grp2 = dest_grp.get_group('SLC')
#Get the counts of each carrier that flies to slc and return then number of rows of carriers that fly to SLC
intq_2 = dest_grp2['carrier'].value_counts().shape[0]
intq_2
```

```python
#Question 3 What is the average arrival delay for flights to RDU? floatq_3

#Group by dest
dest_grp = flights_data.groupby(['dest'])
#Get just the data that flys to RDU
dest_grp2 = dest_grp.get_group('RDU')
#Get the mean of that data
floatq_3 = dest_grp2['arr_delay'].mean()
floatq_3
```

```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  floatq_4 

#filter to rows where the destination is SEA
q4 = flights_data.loc[(flights_data['dest'] == "SEA")]

#summarize the above by origin. Normalize returns the counts as percentages
q4_pt2 = q4['origin'].value_counts(normalize=True)

#use iloc so I can go by integer location to just print the pertentage of flights that came from JFK. There are none from LGA. 
floatq_4 = q4_pt2.iloc[0]
floatq_4
```

```python
#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)q_5 


#Creating a date column 

flights_data['year'] = flights_data['year'].astype(str)
flights_data['month'] = flights_data['month'].astype(str)
flights_data['day'] = flights_data['day'].astype(str)
flights_data['dep_dt'] = flights_data[['year', 'month', 'day']].agg('/'.join, axis=1)


#Group by dep_dt
dep_dt_grp = flights_data.groupby(['dep_dt'])

#average and max 
avg = dep_dt_grp['dep_delay'].mean()
print(avg.idxmax(), ',', avg.max() )
```

```python
#Question 6 Which date has the largest average arrival delay? pd slice with date and floatq_6 
#Group by dep_dt
arr_dt_grp = flights_data.groupby(['dep_dt'])

#average and max 
avg = arr_dt_grp['arr_delay'].mean()
print(avg.idxmax(), ',', avg.max() )
```
