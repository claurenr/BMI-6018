```python
import pandas as pd
import numpy as np
```

```python
breast_cancer_df = pd.read_csv('breast_cancer_data.csv')
breast_cancer_df.head()
```

```python
#1. Melt
#id_vars: tuple, list, or ndarray, Column(s) to use as identifier variables.
#value_vars tuple, list, or ndarray, Column(s) to unpivot. 
#var_name scalar Name to use for the â€˜variableâ€™ column. If None it uses frame.columns.name or â€˜variableâ€™.
#value_name scalar, Name to use for the â€˜valueâ€™ column.

df_melt = breast_cancer_df.melt(id_vars = ['id', 'diagnosis'], value_vars=['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 
                                                              'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean'],
                     var_name = 'Means', value_name = 'Value')

df_melt
```

```python
#2. Pivot
#index[ndarray] : Labels to use to make new frameâ€™s index
#columns[ndarray] : Labels to use to make new frameâ€™s columns
#values[ndarray] : Values to use for populating new frameâ€™s values

df_pivot = df_melt.pivot(index = ['id', 'diagnosis'], columns = 'Means', values = 'Value')
df_pivot
```

```python
#3 Aggregation
#DataFrame.aggregate(func=None, axis=0, *args, **kwargs)
#df.agg(['sum', 'min'])

breast_cancer_df.texture_worst.agg(['min', 'max'])
```

```python
#4. Iteration
for row in df_pivot.itertuples():
    print(row)
```

```python
#5. Groupby

breast_cancer_df.groupby(breast_cancer_df["diagnosis"]).area_worst.agg([ "count",
                                                "min", 
                                               "max", 
                                               "mean"]) 
```
