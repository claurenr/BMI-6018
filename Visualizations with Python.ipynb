```python
#%% viz 3
'''
Create a visualization that shows BOTH the running total of cases for a single
county AND the daily new cases. The graphic should:
-Use two y-axes (https://matplotlib.org/stable/gallery/subplots_axes_and_figures/two_scales.html)
-Use color to contrast the two series being plotted
-Have well formatted dates as the X axis
'''
columns_to_drop = ['iso2','iso3', 'code3', 'FIPS', 'Country_Region', 'Lat', 'Long_', 'Combined_Key', 'Province_State']
cleaned_data_q3 = utah_df.drop(columns = columns_to_drop)
utah_melt_data_q3 = pd.melt(cleaned_data_q3, id_vars=['Admin2'], var_name='Date', value_name='Cases')

group_data_ut_q3 = utah_melt_data_q3.groupby(['Date']).sum('Cases')

```

```python

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

data1 = group_data_ut_q3

# Seaborn scatter plot on the left y-axis
sns.scatterplot(data=data1, x='Date', y='Cases', color='blue', ax=ax1)

# Create a second y-axis on the right
ax2 = ax1.twinx()
data2 = utah_melt_data_q3
# Seaborn scatter plot on the right y-axis
sns.scatterplot(data=data2.loc[(data2['Admin2'] == "Salt Lake")], x='Date', y='Cases', color='green', ax=ax2, marker='s')

# Customize the plot
ax1.set_title('Utah Running total and Daily New Cases for Salt Lake County')
ax1.set_xlabel('Dates')
ax1.set_ylabel('Daily Totals for all Utah Counties', color='blue')
ax2.set_ylabel('Salt Lake County Daily New Cases', color='green')


# Show the plot
plt.show()
```
